```python
import datetime as dt
import pandas as pd
import scipy as s
import numpy as np
import names
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# MANUALLY SELECT: If you choose 2, the program will randomly select
# 10 citations from article 1, and 10 citations from each of those citations.
numIter = 2 

# MANUALLY SELECT: how many random articles to select from each article 
randArticleNum = 10 

# total articles will be the original article * 10 randomly selected 
# from that article * 10 randomly selected from those articles 
totArticles = randArticleNum**(numIter) + 1 

iterArticles = range(0,totArticles) # create range of articles for for loop later

iterArticles[2] # check to see if range worked 
totArticles
```

```python
# create dataframe with # columns include: authors, article title, DOI, and number of article
                        # rows include:    total number of articles 

# for authors, i used a random name generator because it was kind of fun & holds the place for string values later.
# for title, I simply labeled title 1, 2, etc. to be a filler for actual article titles later.
# I simply labeled DOI with what it's respective value will be as well.
# Number is labeled as #.# where the first number will indicate what article the citation is coming from. 
# For example, 0.1 would indicate the original article (0) and the 1st random article (1)

article = pd.DataFrame({"authors": [names.get_full_name()],"title": ["title %d" % (iterArticles[0])], \
                        "DOI": ["DOI%d" % (iterArticles[0])], 'number': [np.nan]})

# print to make sure we did everything right
print(article)        
```

```python
for x in range(1,totArticles):
    print("x = ", x)
    articleAppend = pd.DataFrame({"authors": [names.get_full_name()],"title": ["title %d" % (iterArticles[x])], \
                 "DOI": ["DOI%d" % (iterArticles[x])], 'number': [np.nan]})
    article = article.append(articleAppend, ignore_index = True)

article
```

```python
# now let's assign numbers to each article. 
# Note that article 0.0 = original article
                  # 0.1 to 1.0 = references 1-10 from original article
                  # 1.1 to 2.0 = 1st-10th reference from paper 0.1
                  # 2.1 to 3.0 = 1st-10th reference from paper 0.2
                  # etc. 

article['number'] = np.arange(0.0,totArticles/10,0.1)
article # see how everything looks! 
```

```python
# combine author, article title, and DOI column so they all appear on graph 
article['combined'] = article[['authors', 'title','DOI']].apply(lambda x: '. '.join(x), axis=1)
article
```

```python
## play around with recursion
DG = nx.DiGraph() # initialize networkx graph

DOI = tuple(article.combined[article.number == 0])

def get_data(DOI):
    # go get data from DOI
    data_from_node = {DOI:1}
    return data_from_node

def find_children(DOI):
    # get children DOIs into list from .csv
    article_list = tuple(article.combined[article.number <= 1]) # get DOI children
    return (article_list)

def recursive_add(DOI):
    if not DG.has_node( DOI ): # if there is no node for this DOI
        node_data = get_data(DOI) 
        DG.add_node(DOI)
        child_list = find_children( DOI )
        for child in child_list:
            recursive_add(DOI)
            print( 'adding: '+child )
            DG.add_edge(DOI, child)
            
# def find_childrens_children(DOI)
#     child_list = find_children(DOI)
#     for child in child_list
#         if article.number[article.number == child/10]
#             tuple(article.combined[article.number = 1 and article.number < 2])
#             DG.add_edge(child_list[child],)
        
#     child_list = find_children(DOI)
#     for child in child_list
#         DG.add_edge(child_list[child],)
        
        
            
```
